```python
from data import Database
from altair import Chart
```

```python
# Create dataframe from our database
db = Database()
df = db.dataframe()
cols = ['Level', 'Rarity', 'Health', 'Energy', 'Sanity']
```

```python
source = df.groupby(by='Level').mean().reset_index()

chart1 = Chart(
    source,
    title='Level v. Average Health'
).mark_line().encode(
    x='Level',
    y='Health'
)

chart1
```

```python
chart2 = Chart(
    df[cols],
    title='Level v. Energy for each Rank'
).mark_line().encode(
    x='Level',
    y='Energy',
    color='Rarity'
)

chart2
```

```python
source = df['Rarity'].value_counts().reset_index()

chart3 = Chart(
    source,
    title='Pie Chart of Rarity Counts'
).mark_arc().encode(
    theta='Rarity',
    color='index'
)

chart3
```

```python
source = df.groupby(by='Rarity').mean().reset_index()

chart4 = Chart(
    source,
    title='Rarity v. Average Health'
).mark_line(point=True).encode(
    x='Rarity',
    y='Health'
).properties(
    width=400
)

chart4
```
